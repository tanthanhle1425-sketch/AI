```python
import math
#5x5
SIZE = 5
WIN_CON = 4 # Có 4 quân liên tiếp là thắng (tăng tốc độ game)
MAX_DEPTH = 4 # AI nhìn xa 4 nước

board = [[' ' for _ in range(SIZE)] for _ in range(SIZE)]

def check_winner(b):
    for r in range(SIZE):
        for c in range(SIZE):
            if b[r][c] == ' ': continue
            for dr, dc in [(0,1), (1,0), (1,1), (1,-1)]:
                count = 0
                for i in range(WIN_CON):
                    nr, nc = r + dr*i, c + dc*i
                    if 0 <= nr < SIZE and 0 <= nc < SIZE and b[nr][nc] == b[r][c]:
                        count += 1
                    else: break
                if count == WIN_CON: return b[r][c]
    if all(cell != ' ' for row in b for cell in row): return 'Tie'
    return None

def evaluate_board(b):
    """ Hàm lượng giá: Chấm điểm dựa trên các chuỗi quân cờ """
    # AI sẽ ưu tiên các ô có nhiều quân cờ của mình đứng gần nhau
    score = 0
    # Đây là nơi bạn có thể thêm logic chấm điểm cho chuỗi 2, chuỗi 3
    return score

def minimax(board, depth, alpha, beta, is_max):
    result = check_winner(board)
    if result == 'X': return 100 - depth
    if result == 'O': return depth - 100
    if result == 'Tie': return 0
    if depth >= MAX_DEPTH: return evaluate_board(board)

    if is_max:
        best = -math.inf
        for r in range(SIZE):
            for c in range(SIZE):
                if board[r][c] == ' ':
                    board[r][c] = 'X'
                    score = minimax(board, depth + 1, alpha, beta, False)
                    board[r][c] = ' '
                    best = max(best, score)
                    alpha = max(alpha, score)
                    if beta <= alpha: break # Cắt tỉa
        return best
    else:
        best = math.inf
        for r in range(SIZE):
            for c in range(SIZE):
                if board[r][c] == ' ':
                    board[r][c] = 'O'
                    score = minimax(board, depth + 1, alpha, beta, True)
                    board[r][c] = ' '
                    best = min(best, score)
                    beta = min(beta, score)
                    if beta <= alpha: break # Cắt tỉa
        return best

def find_best_move(b):
    best_val = -math.inf
    move = (-1, -1)
    # Ưu tiên duyệt từ tâm bàn cờ ra ngoài để Alpha-Beta cắt tỉa tốt hơn
    for r in range(SIZE):
        for c in range(SIZE):
            if b[r][c] == ' ':
                b[r][c] = 'X'
                val = minimax(b, 0, -math.inf, math.inf, False)
                b[r][c] = ' '
                if val > best_val:
                    best_val = val
                    move = (r, c)
    return move

def print_board(b):
    print("\n    0   1   2   3   4")
    for i, row in enumerate(b):
        print(f"{i} | " + " | ".join(row) + " |")

# --- Vòng lặp trò chơi ---
print(f"TIC-TAC-TOE 5x5 (Cắt tỉa Alpha-Beta)")
while True:
    print_board(board)
    if check_winner(board): break

    # Lượt người
    try:
        r, c = map(int, input("Nhập hàng và cột (0-4): ").split())
        if board[r][c] != ' ': continue
        board[r][c] = 'O'
    except: continue

    # Lượt AI
    if not check_winner(board):
        print("AI đang tính toán...")
        m = find_best_move(board)
        board[m[0]][m[1]] = 'X'

winner = check_winner(board)
print_board(board)
print(f"\nKẾT THÚC: {winner if winner != 'Tie' else 'Hòa'}")
```

```python
import math
#10x10
SIZE = 10
WIN_CON = 5
DEPTH_LIMIT = 2

board = [[' ' for _ in range(SIZE)] for _ in range(SIZE)]

def check_winner(b):
    """ Kiểm tra xem đã có ai đạt đủ 5 quân liên tiếp chưa """
    for r in range(SIZE):
        for c in range(SIZE):
            if b[r][c] == ' ': continue

            for dr, dc in [(0, 1), (1, 0), (1, 1), (1, -1)]:
                count = 0
                for i in range(WIN_CON):
                    nr, nc = r + dr * i, c + dc * i
                    if 0 <= nr < SIZE and 0 <= nc < SIZE and b[nr][nc] == b[r][c]:
                        count += 1
                    else: break
                if count == WIN_CON: return b[r][c]

    if all(cell != ' ' for row in b for cell in row): return 'Tie'
    return None

def evaluate_line(line, player):
    score = 0
    opp = 'O' if player == 'X' else 'X'
    count = line.count(player)
    empty = line.count(' ')

    if count == 5: score += 100000
    elif count == 4 and empty == 1: score += 10000
    elif count == 3 and empty == 2: score += 1000
    return score

def heuristic_evaluation(b):
    score = 0
    for i in range(SIZE):
        for j in range(SIZE - WIN_CON + 1):
            # Hàng ngang
            row_win = b[i][j:j+WIN_CON]
            score += evaluate_line(row_win, 'X')
            score -= evaluate_line(row_win, 'O')
            # Cột dọc
            col_win = [b[j+k][i] for k in range(WIN_CON)]
            score += evaluate_line(col_win, 'X')
            score -= evaluate_line(col_win, 'O')
    return score

def get_nearby_moves(b):
    moves = set()
    for r in range(SIZE):
        for c in range(SIZE):
            if b[r][c] != ' ':
                for dr in range(-1, 2):
                    for dc in range(-1, 2):
                        nr, nc = r + dr, c + dc
                        if 0 <= nr < SIZE and 0 <= nc < SIZE and b[nr][nc] == ' ':
                            moves.add((nr, nc))
    return list(moves) if moves else [(SIZE//2, SIZE//2)]

def minimax(board, depth, alpha, beta, is_max):

    res = check_winner(board)
    if res == 'X': return 1000000
    if res == 'O': return -1000000
    if res == 'Tie': return 0

    if depth == DEPTH_LIMIT:
        return heuristic_evaluation(board)

    moves = get_nearby_moves(board)
    if is_max:
        best = -math.inf
        for r, c in moves:
            board[r][c] = 'X'
            score = minimax(board, depth + 1, alpha, beta, False)
            board[r][c] = ' '
            best = max(best, score)
            alpha = max(alpha, best)
            if beta <= alpha: break
        return best
    else:
        best = math.inf
        for r, c in moves:
            board[r][c] = 'O'
            score = minimax(board, depth + 1, alpha, beta, True)
            board[r][c] = ' '
            best = min(best, score)
            beta = min(beta, best)
            if beta <= alpha: break
        return best

def find_best_move(b):
    best_val = -math.inf
    move = None
    for r, c in get_nearby_moves(b):
        b[r][c] = 'X'
        val = minimax(b, 0, -math.inf, math.inf, False)
        b[r][c] = ' '
        if val > best_val:
            best_val = val
            move = (r, c)
    return move

def print_board(b):
    print("\n   " + " ".join(f"{i:2}" for i in range(SIZE)))
    for i, row in enumerate(b):
        print(f"{i:2} " + " | ".join(row))

# --- Vòng lặp chính ---
print(f"TIC-TAC-TOE {SIZE}x{SIZE} (Luật 5 quân thắng)")
while True:
    print_board(board)

    # Lượt người
    while True:
        try:
            inp = input("\nLượt bạn (hàng cột): ")
            r, c = map(int, inp.split())
            if board[r][c] == ' ':
                board[r][c] = 'O'
                break
            else: print("Ô đã chọn!")
        except: print("Nhập sai định dạng!")
    winner = check_winner(board)
    if winner: break

    # Lượt AI
    print("AI đang tính toán...")
    m = find_best_move(board)
    if m: board[m[0]][m[1]] = 'X'
    winner = check_winner(board)
    if winner: break

print_board(board)
if winner == 'Tie':
    print("\nKẾT QUẢ: HÒA!")
else:
    print(f"\nKẾT QUẢ: {winner} THẮNG!")
```
