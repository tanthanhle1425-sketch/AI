```python
import math

# Khởi tạo bàn cờ
board = [[' ' for _ in range(3)] for _ in range(3)]

def check_winner(b):
    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] != ' ': return b[i][0]
        if b[0][i] == b[1][i] == b[2][i] != ' ': return b[0][i]
    if b[0][0] == b[1][1] == b[2][2] != ' ': return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] != ' ': return b[0][2]
    if all(cell != ' ' for row in b for cell in row): return 'Tie'
    return None

def minimax(board, depth, alpha, beta, is_maximizing):
    result = check_winner(board)
    if result == 'X': return 10 - depth
    if result == 'O': return depth - 10
    if result == 'Tie': return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    score = minimax(board, depth + 1, alpha, beta, False)
                    board[i][j] = ' '
                    best_score = max(score, best_score)
                    alpha = max(alpha, score)
                    if beta <= alpha: break
        return best_score
    else:
        best_score = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    score = minimax(board, depth + 1, alpha, beta, True)
                    board[i][j] = ' '
                    best_score = min(score, best_score)
                    beta = min(beta, score)
                    if beta <= alpha: break
        return best_score

def find_best_move(board):
    best_val = -math.inf
    move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'
                move_val = minimax(board, 0, -math.inf, math.inf, False)
                board[i][j] = ' '
                if move_val > best_val:
                    best_val = move_val
                    move = (i, j)
    return move

def print_board(b):
    print("\n  0   1   2")
    for i, row in enumerate(b):
        print(f"{i} " + " | ".join(row))
        if i < 2: print("  ---------")

# --- Vòng lặp chính ---
print("TIC-TAC-TOE: BẠN (O) VS AI (X)")
while True:
    print_board(board)

    # 1. Lượt của người chơi
    while True:
        try:
            user_input = input("\nNhập hàng và cột (VD: 1 1): ")
            r, c = map(int, user_input.split())
            if board[r][c] == ' ':
                board[r][c] = 'O'
                break
            else:
                print("Ô này đã có người đi, vui lòng chọn ô khác!")
        except:
            print("Lỗi! Vui lòng nhập 2 số từ 0-2 (ví dụ: 0 2)")

    # 2. Kiểm tra xem người thắng chưa
    res = check_winner(board)
    if res:
        break

    # 3. Lượt của AI
    print("\nAI đang suy nghĩ...")
    m = find_best_move(board)
    if m != (-1, -1):
        board[m[0]][m[1]] = 'X'

    # 4. Kiểm tra xem AI thắng chưa
    res = check_winner(board)
    if res:
        break

# Kết thúc game
print_board(board)
if res == 'Tie':
    print("\nKẾT QUẢ: HÒA!")
else:
    print(f"\nKẾT QUẢ: {res} THẮNG!")
```
