```python
import math

board = [[' ' for _ in range(3)] for _ in range(3)]

def print_board(board):
    for i, row in enumerate(board):
        print(" | ".join(row))
        if i < 2:
            print("-" * 9)

def check_winner(b):

    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] != ' ': return b[i][0]
        if b[0][i] == b[1][i] == b[2][i] != ' ': return b[0][i]

    if b[0][0] == b[1][1] == b[2][2] != ' ': return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] != ' ': return b[0][2]
    if all(cell != ' ' for row in b for cell in row):
        return 'Tie'
    return None

def minimax(board, depth, is_maximizing):
    result = check_winner(board)
    if result == 'X': return 10 - depth
    if result == 'O': return depth - 10
    if result == 'Tie': return 0

    if is_maximizing:
        best_score = -math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'X'
                    score = minimax(board, depth + 1, False)
                    board[i][j] = ' '
                    best_score = max(score, best_score)
        return best_score
    else:
        best_score = math.inf
        for i in range(3):
            for j in range(3):
                if board[i][j] == ' ':
                    board[i][j] = 'O'
                    score = minimax(board, depth + 1, True)
                    board[i][j] = ' '
                    best_score = min(score, best_score)
        return best_score

def best_move(board):
    best_score = -math.inf
    move = None
    for i in range(3):
        for j in range(3):
            if board[i][j] == ' ':
                board[i][j] = 'X'
                score = minimax(board, 0, False)
                board[i][j] = ' '
                if score > best_score:
                    best_score = score
                    move = (i, j)
    return move
print("Chào mừng đến với Tic-Tac-Toe AI!")
print("Bạn là 'O', Máy là 'X'. Nhập tọa độ hàng và cột (0, 1, 2).")

while True:
    print_board(board)
    winner = check_winner(board)
    if winner:
        if winner == 'Tie': print("\nKết quả: Hòa!")
        else: print(f"\nKết quả: {winner} thắng!")
        break


    try:
        row, col = map(int, input("Lượt của bạn (hàng cột): ").split())
        if board[row][col] != ' ':
            print("Ô này đã có người đi. Thử lại!")
            continue
        board[row][col] = 'O'
    except (ValueError, IndexError):
        print("Vui lòng nhập 2 số từ 0 đến 2 (VD: 1 1).")
        continue


    move = best_move(board)
    if move:
        board[move[0]][move[1]] = 'X'
        print(f"\nMáy đi vào ô: {move[0]} {move[1]}")
```
