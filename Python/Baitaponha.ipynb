```python
# Nhập các phần tử cho mảng a
arr_a_input = input("Nhập các phần tử của mảng a, cách nhau bởi dấu cách : ")
arr_a = [int(x) for x in arr_a_input.split()]

# Nhập các phần tử cho mảng b
arr_b_input = input("Nhập các phần tử của mảng b, cách nhau bởi dấu cách  ")
arr_b = [int(x) for x in arr_b_input.split()]

print(f"Mảng a bạn đã nhập: {arr_a}")
print(f"Mảng b bạn đã nhập: {arr_b}")
def merge_arrays_with_min(arr1, arr2):
    result = []
    len1 = len(arr1)
    len2 = len(arr2)
    min_len = min(len1, len2)

    # Merge elements where both arrays have corresponding values
    for i in range(min_len):
        result.append(min(arr1[i], arr2[i]))

    # Append remaining elements from arr1, if any
    if len1 > min_len:
        result.extend(arr1[min_len:])

    # Append remaining elements from arr2, if any
    if len2 > min_len:
        result.extend(arr2[min_len:])

    return result
    # Gọi hàm merge_arrays_with_min với các mảng vừa nhập
merged_array_dynamic = merge_arrays_with_min(arr_a, arr_b)
print(f"Mảng kết quả: {merged_array_dynamic}")
# Ví dụ sử dụng:
#arr_a_example = [3, 9, 1, 4]
#arr_b_example = [2, 7, 4, 3, 2, 8]

#merged_array_example = merge_arrays_with_min(arr_a_example, arr_b_example)
#print(f"Mảng a: {arr_a_example}")
#print(f"Mảng b: {arr_b_example}")
#print(f"Mảng kết quả: {merged_array_example}")
```

```python
# Nhập số dòng m
m = int(input("Nhập số dòng (m) cho mảng 2 chiều: "))

# Nhập số cột n
n = int(input("Nhập số cột (n) cho mảng 2 chiều: "))

# Khởi tạo mảng 2 chiều rỗng
array_2d = []

print(f"Nhập các phần tử cho mảng {m}x{n}:")

# Nhập các phần tử cho từng dòng và cột
for i in range(m):
    row = []
    for j in range(n):
        element = int(input(f"Nhập phần tử tại vị trí [{i}][{j}]: "))
        row.append(element)
    array_2d.append(row)

print("Mảng 2 chiều bạn đã nhập là:")
for row in array_2d:
    print(row)
sum_upper_anti_diagonal = 0

# Kiểm tra xem mảng có phải là ma trận vuông không
# Số dòng (m) và số cột (n) đã được nhập từ cell trước.
# Nếu m != n, thì khái niệm 'tam giác trên của đường chéo phụ' không rõ ràng.
# Tuy nhiên, dựa trên yêu cầu 'ma trận vuông a cấp n', chúng ta sẽ giả định m = n.

if m != n:
    print("Mảng bạn nhập không phải là ma trận vuông. Không thể tính tam giác trên của đường chéo phụ.")
else:
    for i in range(m):
        for j in range(n):
            # Điều kiện i + j <= n - 1 xác định các phần tử trên và bao gồm đường chéo phụ
            if i + j <= n - 1:
                sum_upper_anti_diagonal += array_2d[i][j]
    print(f"Tổng các phần tử thuộc tam giác trên của đường chéo phụ (kể cả đường chéo phụ) là: {sum_upper_anti_diagonal}")
array_2d_abs = [[abs(element) for element in row] for row in array_2d]

print("Mảng 2 chiều sau khi chuyển các phần tử âm thành trị tuyệt đối:")
for row in array_2d_abs:
    print(row)
x = int(input("Nhập số nguyên x để thay thế các phần tử chẵn: "))

array_2d_replaced = []
for i in range(m):
    new_row = []
    for j in range(n):
        if array_2d[i][j] % 2 == 0:  # Kiểm tra phần tử chẵn
            new_row.append(x)
        else:
            new_row.append(array_2d[i][j])
    array_2d_replaced.append(new_row)

print(f"Mảng 2 chiều sau khi thay thế các phần tử chẵn bằng {x}:")
for row in array_2d_replaced:
    print(row)
all_even = True
for row in array_2d:
    for element in row:
        if element % 2 != 0:  # Nếu tìm thấy phần tử lẻ
            all_even = False
            break
    if not all_even:
        break

if all_even:
    print("Mảng 2 chiều 'array_2d' chứa toàn bộ các phần tử chẵn.")
else:
    print("Mảng 2 chiều 'array_2d' có chứa ít nhất một phần tử lẻ.")
def is_symmetric_matrix(matrix):
    """
    Kiểm tra xem ma trận vuông có đối xứng hay không.
    Một ma trận đối xứng khi a[i][j] = a[j][i] cho mọi i, j.
    """
    n = len(matrix)
    # Kiểm tra nếu không phải ma trận vuông, mặc dù đề bài giả định là ma trận vuông.
    for row in matrix:
        if len(row) != n:
            print("Đây không phải là ma trận vuông.")
            return False

    for i in range(n):
        for j in range(i + 1, n):  # Chỉ cần kiểm tra các phần tử trên đường chéo chính
            if matrix[i][j] != matrix[j][i]:
                return False
    return True

# Kiểm tra với ma trận array_2d đã có
print(f"Ma trận của bạn: {array_2d}")
if is_symmetric_matrix(array_2d):
    print("Ma trận là đối xứng.")
else:
    print("Ma trận không đối xứng.")
def is_main_diagonal_increasing(matrix):
    """
    Kiểm tra xem đường chéo chính của ma trận vuông có tăng dần không.
    """
    n = len(matrix)
    # Đảm bảo là ma trận vuông
    for row in matrix:
        if len(row) != n:
            print("Đây không phải là ma trận vuông.")
            return False

    if n <= 1:  # Ma trận 0x0 hoặc 1x1 luôn có đường chéo chính tăng dần (hoặc không có gì để so sánh)
        return True

    # Lấy các phần tử trên đường chéo chính và kiểm tra
    for i in range(1, n):
        if matrix[i][i] <= matrix[i-1][i-1]:
            return False
    return True

# Kiểm tra với ma trận array_2d đã có
if is_main_diagonal_increasing(array_2d):
    print("Đường chéo chính của ma trận là tăng dần.")
else:
    print("Đường chéo chính của ma trận không tăng dần.")
lower_anti_diagonal_elements = []
# Đối với ma trận vuông, m = n. Sử dụng n để duyệt.
# Điều kiện cho tam giác dưới của đường chéo phụ là i + j >= n - 1
for i in range(m):
    for j in range(n):
        if i + j >= n - 1:
            lower_anti_diagonal_elements.append(array_2d[i][j])

print(f"Các phần tử thuộc tam giác dưới của đường chéo phụ (kể cả đường chéo phụ) là: {lower_anti_diagonal_elements}")
def is_anti_diagonal_decreasing(matrix):
    """
    Kiểm tra xem đường chéo phụ của ma trận vuông có giảm dần hay không.
    """
    n = len(matrix)
    # Đảm bảo là ma trận vuông
    for row in matrix:
        if len(row) != n:
            print("Đây không phải là ma trận vuông.")
            return False

    if n <= 1:  # Ma trận 0x0 hoặc 1x1 luôn coi là giảm dần (hoặc không có gì để so sánh)
        return True

    # Lấy các phần tử trên đường chéo phụ và kiểm tra
    # Các phần tử đường chéo phụ có chỉ số (i, n-1-i)
    for i in range(1, n):
        # So sánh phần tử hiện tại với phần tử trước đó trên đường chéo phụ
        if matrix[i][n - 1 - i] >= matrix[i-1][n - 1 - (i-1)]:
            return False
    return True

# Kiểm tra với ma trận array_2d đã có
if is_anti_diagonal_decreasing(array_2d):
    print("Đường chéo phụ của ma trận là giảm dần.")
else:
    print("Đường chéo phụ của ma trận không giảm dần.")
```
